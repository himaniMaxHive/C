# Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
res = classifier("I was so happy with today's task completion, as it was nearly impossible to complete")
print(res)
```

```python
pipeline(task = 'sentiment-analysis')('I was confused with Barbie movie')
```
